```python
# Deschide notebook-ul și caută apelul render_prompt
# Sau verifică direct dacă template-ul v4 are bloc few-shot

from pathlib import Path

# 1. Verifică template-urile
for folder in ["prompts/incongruities", "prompts", "configs"]:
    p = Path(folder)
    if p.exists():
        for f in sorted(p.rglob("*v4*")):
            print(f)
        for f in sorted(p.rglob("*.j2")):
            content = f.read_text(encoding="utf-8")
            has_fs = "few_shot" in content
            print(f"  {f}  ({len(content)} chars)  few_shot_in_template={has_fs}")

print("\n---\n")

# 2. Caută render_prompt în notebook
import os
for nb in Path(".").rglob("*incongruiti*.ipynb"):
    print(f"Found notebook: {nb}")
    content = nb.read_text(encoding="utf-8")
    
    # Caută toate aparițiile render_prompt
    lines = content.split("\\n")
    for i, line in enumerate(lines):
        if "render_prompt" in line:
            print(f"  Line ~{i}: {line.strip()[:150]}")
```

```python
import json
from pathlib import Path

out_dir = Path("outputs_incongruities")

for f in sorted(out_dir.glob("*.json")):
    with open(f, encoding="utf-8") as fh:
        data = json.load(fh)
    
    # Extrage promptul (adaptează key-ul)
    preds = data.get("predictions", data.get("results", []))
    if preds and "prompt_sent" in preds[0]:
        avg_len = sum(len(str(p["prompt_sent"])) for p in preds) / len(preds)
    else:
        avg_len = 0
    
    name = f.stem
    print(f"{name:<50} avg_prompt_len={avg_len:.0f}")
```
